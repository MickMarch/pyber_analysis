```python
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

stdev = statistics.stdev(y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color= "green", label= "Boston")

# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add a legend
plt.legend()
```

```python
# Create the plot using barh()
plt.barh(x_axis, y_axis, color= "green", label= "Boston")

# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add a legend
plt.legend()
```

```python
# Create the plot using barh() and INVERT the y-axis
plt.barh(x_axis, y_axis, color= "green", label= "Boston")

# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add a legend
plt.legend()

# Invert the y-axis
plt.gca().invert_yaxis()
```

```python
# Create the plot using barh()
plt.barh(x_axis, y_axis, color= "magenta", label= "Boston")

# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add a legend
plt.legend()

# Invert the y-axis
plt.gca().invert_yaxis()
```

## Object Oriented Approach to Bar Charts

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot using the OO approach with barh() and color the cyan.
# Also added error bars
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color= "cyan", label= "Chicago", xerr= stdev, capsize= 3)

# Set title
ax.set_title("PyBer Fare by Month")

# Set x and y labels
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")

# Add legend
ax.legend()

# Invert the y-axis data
ax.invert_yaxis()
```

```python
# Add error bar WITHIN bar(yerr=<value>, capsize=<int>)
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```

# BONUS ERROR BAR

```python
import numpy as np


# example data
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

# example error bar values that vary with x-position
error = 0.1 + 0.2 * x

fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True)
ax0.errorbar(x, y, yerr=error, fmt='-o')
ax0.set_title('variable, symmetric error')

# error bar values w/ different -/+ errors that
# also vary with the x-position
lower_error = 0.4 * error
upper_error = error
asymmetric_error = [lower_error, upper_error]

ax1.errorbar(x, y, xerr=asymmetric_error, fmt='o')
ax1.set_title('variable, asymmetric error')
ax1.set_yscale('log')
plt.show()
```
