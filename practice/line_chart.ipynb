```python
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

stdev = statistics.stdev(y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="x", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
""" same thing as above, but with an error bar """

# Create the plot.
plt.plot(x_axis, y_axis, marker="x", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(-3, 55)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
plt.errorbar(x_axis, y_axis, yerr= stdev, capsize= 3)
# Add the legend.
plt.legend()
```

## Object Oriented Approach

```python
fig, ax = plt.subplots()

ax.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label='Boston')
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,45)
ax.grid(True)
```

```python
""" same thing as above, but with an error bar """

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label='Boston')
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylim(-4,57)
ax.errorbar(x_axis, y_axis, yerr= stdev, capsize=3)
ax.grid(True)
```

```python
""" same thing as above, but with Major and Minor Tick changes """
from matplotlib.ticker import MultipleLocator
import numpy as np

fig, ax = plt.subplots(figsize= (8, 6))

ax.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label='Boston')
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.legend()
ax.set_title("PyBer Fare by Month")

# set Major ticks on the y-axis to step by 5
ax.set_yticks(np.arange(0, 51, step= 5.0))

# set Minor ticks on the y-axis to increments of 1: MultipleLocator()
ax.yaxis.set_minor_locator(MultipleLocator(1))

ax.errorbar(x_axis, y_axis, yerr= stdev, capsize=3)
ax.grid(True)
```
