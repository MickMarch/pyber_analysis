```python
import matplotlib.pyplot as plt
```

## Scatter Plot

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create scatter with plot()

plt.plot(x_axis, y_axis,"o")
```

```python
# Create plot with scatter()

plt.scatter(x_axis, y_axis)
```

```python
# Plot scatter and change the color of the markers to red, and have the fares on the x-axis
plt.scatter(y_axis, x_axis, color= "red", label = "Chicago")

# Add a legend for the city of Chicago.
plt.legend()

# Add a title and axes labels.
plt.title("Pyber Fare by Month")
plt.xlabel("Fares ($)")
plt.ylabel("Date")

# Add a limit for the x-axis data.
plt.xlim(0, 45)

# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

# Bubble Plot

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

## Object Oriented Approach

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Add a legend for the city of Boston.

# Invert the y-axis so that January is at the top.
#-----

# Create bubble plot with scatter(), color = sky blue (20% transparent), linewidth = 2, 
# black edge, points are 5* bigger than the data amount, month values on y-axis
fig, ax = plt.subplots()
ax.scatter(
    y_axis,
    x_axis,
    c= "skyblue",
    linewidths= 2,
    edgecolors= "black",
    alpha= 0.8,
    s= [i*5 for i in y_axis],
)

# Set chart title
ax.set_title("PyBar Fares by Month")

# Set labels for x and y axis
ax.set_ylabel("Date")
ax.set_xlabel("Fares ($)")

# Set x-axis limits
ax.set_xlim(0, 45)

# Add a legend for Boston
ax.legend(["Boston"])

# Invert y-axis so January is at the top

ax.invert_yaxis()

```
