### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os

# File to Load (Remember to check these)
city_data_to_load = os.path.join("resources", "city_data.csv")
ride_data_to_load = os.path.join('resources', 'ride_data.csv')

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
# Display the data table for preview of city_data_df
city_data_df.head()
```

```python
# Display the data table for preview of ride_data_df
ride_data_df.head()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge 
## Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
ride_count_city_type = pyber_data_df.groupby("type").count()["ride_id"]
ride_count_city_type
```

```python
# 2. Get the total drivers for each city type
driver_count_city_type = city_data_df.groupby("type").sum()["driver_count"]
driver_count_city_type
```

```python
#  3. Get the total amount of fares for each city type
fare_count_city_type = pyber_data_df.groupby('type').sum()['fare']
fare_count_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_ride_fare_city_type = fare_count_city_type / ride_count_city_type
avg_ride_fare_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_driver_fare_city_type = fare_count_city_type / driver_count_city_type
avg_driver_fare_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary = pd.DataFrame({
    "Total Rides": ride_count_city_type,
    "Total Drivers": driver_count_city_type, 
    "Total Fares": fare_count_city_type,
    "Average Fare per Ride": avg_ride_fare_city_type,
    "Average Fare per Driver": avg_driver_fare_city_type
})
pyber_summary
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary.index.name = None
pyber_summary
```

```python
#  8. Format the columns.
summary_format = {
    "Total Rides": '{:,}',
    "Total Drivers": '{:,}',
    "Total Fares": '${:,.2f}',
    "Average Fare per Ride": '${:,.2f}',
    "Average Fare per Driver": '${:,.2f}',
}

pyber_summary.style.format(summary_format)

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df.head()
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

fares_by_date = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
fares_by_date.head()
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_by_date = fares_by_date.reset_index()
fares_by_date.head(2)
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

fares_by_date = pd.pivot(fares_by_date, index="date", columns="type", values= "fare")
fares_by_date
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
fares_by_date_filtered = fares_by_date.loc['2019-01-01':'2019-04-28']
fares_by_date_filtered
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fares_by_date_filtered.index = pd.to_datetime(fares_by_date_filtered.index)
fares_by_date_filtered
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fares_by_date_filtered.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = fares_by_date_filtered.resample("W").sum()
weekly_fares_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

weekly_fares = weekly_fares_df.plot()
weekly_fares.set_title("Total Fare by City Type (2019)")
weekly_fares.set_ylabel("Fare($USD)")
weekly_fares.set_xlabel('')
weekly_fares.figure.set_figheight(6)
weekly_fares.figure.set_figwidth(20)

plt.tight_layout()
plt.savefig(os.path.join('analysis', 'pyber_fare_summary.png'))
plt.show()
```
